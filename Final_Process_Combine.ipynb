```python
import csv
import tqdm
```

```python
def read_data(filename):
    with open(filename) as f:
        reader = csv.DictReader(f)
        
        user_id = list()
        time = list()
        
        for row in reader:
            user_id.append(int(row['user_id']))
            time.append(row['timestamp'])
    return user_id, time
```

```python
def change_time(time):
    pre_time = time[0]
    temp_time = 1
    res = list()
    for sub_time in time:
        if sub_time == pre_time:
            sub_time = temp_time
            res.append(int(sub_time))
        else:
            temp_time += 1
            pre_time = sub_time
            sub_time = temp_time
            res.append(int(sub_time))
    return res
```

```python
def extract_data(user_id_data, user_id_label, time_data, time_label):
    length_1 = len(user_id_data)
    length_2 = len(user_id_label)
    index1 = 0
    index2 = 0
    while index1 < length_1 and index2 < length_2:
        #print('index1: ', index1, ' index2: ', index2)
        if time_data[index1] == time_label[index2]:
            if user_id_data[index1] < user_id_label[index2]:
                user_id_data[index1] = 'XXXXX'
                index1 += 1
            elif user_id_data[index1] > user_id_label[index2]:
                user_id_label[index2] = 'XXXXX'
                index2 += 1
            else:
                index1 += 1
                index2 += 1
        elif time_data[index1] > time_label[index2]:
            user_id_label[index2] = 'XXXXX'
            index2 += 1
            
        else:
            user_id_data[index1] = 'XXXXX'
            index1 += 1
        
     
    return user_id_data, user_id_label
```

```python
def main():
    user_id_data, time_data = read_data('data/final_file_temp.csv')
    user_id_label, time_label = read_data('data/label.csv')
    time_data = change_time(time_data)
    time_label = change_time(time_label)
    data_train, data_label = extract_data(user_id_data, user_id_label, time_data, time_label)
    for sub_list in data_label:
        print(sub_list)
    #write(data, filename)
```
